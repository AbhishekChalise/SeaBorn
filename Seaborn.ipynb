```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
crash_df = sns.load_dataset("car_crashes")
crash_df
```

```python
# Distribution Plot

sns.distplot(crash_df['not_distracted'],kde = False, bins =20)
```

```python
# Joint Plot
sns.jointplot(x='alcohol',y='not_distracted',data=crash_df,kind='hex')
```

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
# pair Plot
sns.pairplot(crash_df)
```

```python
tips = sns.load_dataset('tips')
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette ='Blues')
```

```python
#Rug Plot
sns.rugplot(tips['tip'])
```

```python
# Styling the diagrams
sns.set_style('white')
plt.figure(figsize=(8,4))
sns.jointplot(x='speeding', y='alcohol', data=crash_df , kind='reg')
```

```python
# Styling the diagrams
sns.set_style('ticks')
plt.figure(figsize=(8,4))
sns.set_context =('paper',font_scale:=1.4)
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
sns.despine(left=True,bottom=True)
```

```python
sns.barplot(x='sex',y='total_bill',data = tips,estimator = np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
plt.legend(loc = 0)
```

```python
# Violin Plot
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#Strip Plot

plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

```python
# Swarm Plot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='white')
```

```python
# Palettes
plt.figure(figsize=(8,5))
sns.set_style('dark')

sns.stripplot(x='day',y='total_bill',data = tips,hue='sex')
'''==================   =============
    Location String      Location Code
    ==================   =============
    'best' (Axes only)   0
    'upper right'        1
    'upper left'         2
    'lower left'         3
    'lower right'        4
    'right'              5
    'center left'        6
    'center right'       7
    'lower center'       8
    'upper center'       9
    'center'             10
    ==================   ============='''
plt.legend(loc=1)
```

```python


plt.figure(figsize=(8,6))

# Correctly set the context
sns.set_context=('paper', font_scale:=1.4)

# Select only numeric columns from crash_df to avoid string conversion issues
numeric_df = crash_df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
crash_mx = numeric_df.corr()

# Create the heatmap
sns.heatmap(crash_mx, annot=True, cmap='Blues')

plt.show()
```
